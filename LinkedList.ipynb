# 1. Get Node Value for Node 'n' places from the end

### Solution: Initialize two pointers. Move pointer1, n places from the start. Then start moving pointer2. pointer1 will reach the end, while pointer2 will be n places from the end

```python
#!/bin/python3

import math
import os
import random
import re
import sys
from random import randint

class SinglyLinkedListNode:
    def __init__(self, node_data):
        self.data = node_data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_node(self, node_data):
        node = SinglyLinkedListNode(node_data)

        if not self.head:
            self.head = node
        else:
            self.tail.next = node


        self.tail = node

def print_singly_linked_list(node, sep, fptr):
    while node:
        fptr.write(str(node.data))

        node = node.next

        if node:
            fptr.write(sep)

# Complete the getNode function below.

#
# For your reference:
#
# SinglyLinkedListNode:
#     int data
#     SinglyLinkedListNode next
#
#
def getNode(head, positionFromTail):
    ptr1 = head
    ptr2 = head
    count = 0
    
    while ptr1.next and count < positionFromTail:
        ptr1 = ptr1.next
        count += 1

    while ptr1.next:
        ptr1 = ptr1.next
        ptr2 = ptr2.next
    
    return ptr2.data
            

if __name__ == '__main__':
    #fptr = open(os.environ['OUTPUT_PATH'], 'w')

    #tests = int(input())

    #for tests_itr in range(tests):
    llist_count = 10

    llist = SinglyLinkedList()

    for _ in range(llist_count):
        llist_item = randint(0, 22)
        print(llist_item)
        llist.insert_node(llist_item)

    position = 3

    result = getNode(llist.head, position)
    print("Value of Node:")
    print(result)

        #fptr.write(str(result) + '\n')

    #fptr.close()
```

# 2. Delete duplicate-value nodes from a sorted linked list
